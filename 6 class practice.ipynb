### cleaning columns to concat

```python
wh2015.columns = wh2015.columns.str.replace("("," ").str.replace(")"," ").str.strip().str.upper()
wh2015.columns
```

```python
wh2016.columns = wh2016.columns.str.replace("("," ").str.replace(")"," ").str.strip().str.upper()
wh2016.columns
```

```python
 wh2017.columns = wh2017.columns.str.replace("."," ").str.replace("\s+"," ").str.strip().str.upper()
wh2017.columns
```

```python
combine = pd.concat([wh2015,wh2016,wh2017], ignore_index = True)
combine
```

```python
combine.isnull().sum()
```

### graphical representation

```python
import seaborn as sns
```

```python
combine.set_index("YEAR")
```

```python
sns.heatmap(combine.set_index("YEAR").isnull())
plt.show
```

### filling missing values in region columns

```python
combine = pd.merge(combine,region, on = "COUNTRY" , how = "inner")
```

```python
combine.isnull().sum()
```

```python
combine = combine.drop("REGION_x" , axis = 1)
```

```python
combine
```

### changing location of region column

```python
region_column = combine.pop("REGION_y")
```
